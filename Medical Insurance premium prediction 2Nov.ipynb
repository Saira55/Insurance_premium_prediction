```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load your dataset
# insurance_df = pd.read_csv('path_to_your_dataset.csv')

# Custom function to display correlation coefficients with asterisks for high correlations
def cor_annot(x, y, **kws):
    r, _ = pearsonr(x, y)
    ax = plt.gca()
    ax.annotate(f"{r:.2f}{'**' if abs(r) > 0.6 else ''}", xy=(.1, .9), xycoords=ax.transAxes, color='#65235B', size=25)

# Function to calculate correlation only for numerical pairs
def cor_pair(grid):
    for i, j in zip(*np.triu_indices_from(grid.axes, 1)):
        grid.axes[j, i].scatter(grid.x_vars[i], grid.y_vars[j])
        cor_annot(grid.data[grid.x_vars[i]], grid.data[grid.y_vars[j]])

# Creating the pairplot with customization
pair_plot = sns.PairGrid(insurance_df)
pair_plot.map_lower(sns.scatterplot, color='#65235B')
pair_plot.map_upper(cor_annot)

pair_plot.map_diag(sns.kdeplot, lw=3, legend=False, color='#65235B')

# Removing background color and grid lines globally
for ax in pair_plot.axes.flatten():
    ax.set_facecolor('white')  # White background
    ax.grid(False)  # Remove gridlines

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, accuracy_score

regression_model_sklearn=LinearRegression()
regression_model_sklearn.fit(X_train, y_train)
```

```python
k=X_test.shape[1]
n=len(X_test)
n
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt
```

```python
RMSE=float(format(np.sqrt(mean_squared_error(y_test_orig,y_predict_orig)), '.3f'))
MSE=mean_squared_error(y_test_orig,y_predict_orig)
MAE=mean_absolute_error(y_test_orig,y_predict_orig)
r2=r2_score(y_test_orig,y_predict_orig)
adj_r2=1-(1-r2)*(n-1)/(n-k-1)

print('RMSE=', RMSE, '\nMSE=', MSE, '\nMAE=', MAE, '\nR2=',r2, '\nAdjusted R2=', adj_r2)
```
